```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/Users/vamsikeshwaranm/Downloads/CTS/monthly_corrected_no_outliers.csv")
df['datum'] = pd.to_datetime(df['datum'])

results = []
```

```python
param_grid = [
    {"yearly_seasonality": True, "weekly_seasonality": False, "daily_seasonality": False, "changepoint_prior_scale": 0.05},
    {"yearly_seasonality": True, "weekly_seasonality": True, "daily_seasonality": False, "changepoint_prior_scale": 0.5},
    {"yearly_seasonality": False, "weekly_seasonality": False, "daily_seasonality": True, "changepoint_prior_scale": 0.1},
    {"yearly_seasonality": True, "weekly_seasonality": False, "daily_seasonality": False, "changepoint_prior_scale": 1.0},
]
```

```python
for col in df.columns:
    if col == 'datum':
        continue
    
    print(f"\nProcessing column: {col}")
    sample = df[['datum', col]].rename(columns={'datum': 'ds', col: 'y'})
    
    best_model, best_forecast, best_acc, best_metrics = None, None, -1, None
    
    for params in param_grid:
        try:
            m = Prophet(interval_width=0.95, **params)
            m.fit(sample)
            
            future = m.make_future_dataframe(periods=10, freq='M')
            forecast = m.predict(future)
            
            merged = pd.merge(sample, forecast, on='ds', how='inner')
            actual = merged['y'].values
            predicted = merged['yhat'].values
            
            mape = mean_absolute_percentage_error(actual, predicted) * 100
            rmse = np.sqrt(mean_squared_error(actual, predicted))
            r2 = r2_score(actual, predicted)
            accuracy = 100 - mape
            
            if accuracy > best_acc:  
                best_model, best_forecast = m, forecast
                best_acc = accuracy
                best_metrics = (mape, rmse, r2)
        
        except Exception as e:
            print(f"Skipping config {params} due to error: {e}")
            continue
            
    # Store results
    results.append({
        "Column": col,
        "Accuracy (%)": round(best_acc, 2),
        "MAPE (%)": round(best_metrics[0], 2),
        "RMSE": round(best_metrics[1], 2),
        "R²": round(best_metrics[2], 4)
    })

    print(f"\nForecasted values for {col}:")
    print(best_forecast[['ds', 'yhat']].tail(10).to_string(index=False))

    print(f"\nExplainability for {col}:")
    
    fig = best_model.plot_components(best_forecast)
    plt.show()

    changepoints = best_model.changepoints
    deltas = best_model.params["delta"].mean(0) 
    top_changepoints = pd.DataFrame({
        "Date": changepoints,
        "Trend_Change": deltas
    }).sort_values(by="Trend_Change", key=abs, ascending=False).head(5)

    print("Top 5 changepoints (largest trend shifts):")
    print(top_changepoints.to_string(index=False))


    trend_last = best_forecast['trend'].tail(10).values
    if trend_last[-1] > trend_last[0]:
        print(f"\n📈 The overall trend for {col} is increasing. Sales are expected to grow.")
    elif trend_last[-1] < trend_last[0]:
        print(f"\n📉 The overall trend for {col} is decreasing. Sales may decline.")
    else:
        print(f"\n➖ The overall trend for {col} is flat. Sales are expected to stay stable.")

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(sample['ds'], sample['y'], label="Actual", marker='o')
    plt.plot(best_forecast['ds'], best_forecast['yhat'], label="Forecast", marker='x')
    plt.title(f"Forecast vs Actual for {col}\nAccuracy={best_acc:.2f}%, RMSE={best_metrics[1]:.2f}, MAPE={best_metrics[0]:.2f}%")
    plt.xlabel("Date")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
results_df = pd.DataFrame(results)
print("\nFinal Accuracy Results:")
print(results_df)
print(f"Total accuracy: {np.sum(results_df['Accuracy (%)'])/len(results_df['Accuracy (%)'])}")
```
